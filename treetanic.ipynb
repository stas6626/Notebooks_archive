```python
import numpy as np
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
import pandas
```

```python
data = pandas.read_csv('/home/stas/data/titanic.csv', index_col='PassengerId')
```

```python
clf=DecisionTreeClassifier(random_state=241)
```

```python
x=np.array([[1,2],[3,4],[5,6]])
```

```python
y=np.array([0,1,0])
```

```python
clf.fit(x,y)
```

```python
importances=clf.feature_importances_
```

```python
frame=data.loc[:,["Sex","Pclass","Age","Fare"]]
```

```python
%%time
for i in range(1,len(frame["Sex"])+1):
    if frame["Sex"][i]=="male":
        frame["Sex"][i]=0
    else: 
        frame["Sex"][i]=1
```
